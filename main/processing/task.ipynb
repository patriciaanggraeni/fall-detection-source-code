```python
def exclude_data(data, length):
    ldata = joblib.load(data)
    exc = [i for i in ldata if i.shape[0] >= length]

    return exc
```

```python
utask_01 = exclude_data('../../assets/pkl/update/task/task_01.pkl', 600)
utask_06 = exclude_data('../../assets/pkl/update/task/task_06.pkl', 600)
utask_20 = exclude_data('../../assets/pkl/update/task/task_20.pkl', 600)
utask_21 = exclude_data('../../assets/pkl/update/task/task_21.pkl', 600)
utask_22 = exclude_data('../../assets/pkl/update/task/task_22.pkl', 600)

if not os.path.exists('../../assets/pkl/update/utask'):
    os.makedirs('../../assets/pkl/update/utask')
    
    joblib.dump(utask_01, '../../assets/pkl/update/utask/utask01.pkl')
    joblib.dump(utask_06, '../../assets/pkl/update/utask/utask06.pkl')
    joblib.dump(utask_20, '../../assets/pkl/update/utask/utask20.pkl')
    joblib.dump(utask_21, '../../assets/pkl/update/utask/utask21.pkl')
    joblib.dump(utask_22, '../../assets/pkl/update/utask/utask22.pkl')
```

```python
def cut_off(data, start, end, fs=100):
    if start == 0:
        start = 0
    else:
        start = (start * fs)
    end = (end * fs)

    return data[start:end]

def loop_cut_off(data, start, end, fs=100):
    new_data = []
    for i in data:
        new_data.append(cut_off(i, start, end, fs))
    return new_data
```

```python
task_01_cut = loop_cut_off(joblib.load('../../assets/pkl/update/utask/utask01.pkl'), 0, 6)
task_06_cut = loop_cut_off(joblib.load('../../assets/pkl/update/utask/utask06.pkl'), 2, 8)
task_20_cut = loop_cut_off(joblib.load('../../assets/pkl/update/utask/utask20.pkl'), 0, 6)
task_21_cut = loop_cut_off(joblib.load('../../assets/pkl/update/utask/utask21.pkl'), 0, 6)
task_22_cut = loop_cut_off(joblib.load('../../assets/pkl/update/utask/utask22.pkl'), 0, 6)

if not os.path.exists('../../assets/pkl/update/task_cut'):
    os.makedirs('../../assets/pkl/update/task_cut')
    
    joblib.dump(utask_01, '../../assets/pkl/update/task_cut/task_01_cut.pkl')
    joblib.dump(utask_06, '../../assets/pkl/update/task_cut/task_06_cut.pkl')
    joblib.dump(utask_20, '../../assets/pkl/update/task_cut/task_20_cut.pkl')
    joblib.dump(utask_21, '../../assets/pkl/update/task_cut/task_21_cut.pkl')
    joblib.dump(utask_22, '../../assets/pkl/update/task_cut/task_22_cut.pkl')
```
